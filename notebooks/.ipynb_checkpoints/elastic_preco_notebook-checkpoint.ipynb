## 1.1 Rename Columns

```python
# Change to snake case
old_cols = ['Date_imp_d', 'Category_name', 'name', 'price', 'merchant', 'brand',
       'manufacturer', 'Day_n', 'month', 'month_n', 'day', 'Week_Number']

snake_case = lambda x: inflection.underscore(x)

new_cols = list( map(snake_case, old_cols))
df1.columns = new_cols
```

## 1.8 Descriptive Statistical

```python
# Separação entre variáveis numéricas e categóricas
num_attributes = df1.select_dtypes( include=['int64', 'float64'])
cat_attributes = df1.select_dtypes( exclude=['int64', 'float64', 'datetime64[ns]'])
```

### 1.8.1 Numerical Attributes

```python
# Central tendency - mean, median
ct1 = pd.DataFrame(num_attributes.apply(np.mean)).T
ct2 = pd.DataFrame(num_attributes.apply(np.median)).T

# Dispersion - std, max, min, range, skew, kurtosis
d1 = pd.DataFrame(num_attributes.apply(np.std)).T
d2 = pd.DataFrame(num_attributes.apply(min)).T
d3 = pd.DataFrame(num_attributes.apply(max)).T
d4 = pd.DataFrame(num_attributes.apply(lambda x: x.max() - x.min())).T
d5 = pd.DataFrame(num_attributes.apply(lambda x: x.skew())).T
d6 = pd.DataFrame(num_attributes.apply(lambda x: x.kurtosis())).T

# Concatenate
m = pd.concat([d2, d3, d4, ct1, ct2, d1, d5, d6]).T.reset_index()
m.columns = ['attributes', 'min', 'max', 'range', 'mean', 'median', 'std', 'skew', 'kurtosis']

m
```

### 1.8.2 Categorical Atribbutes

```python
# Quantas variáveis categóricas em cada feature
cat_attributes.apply( lambda x: x.unique().shape[0] )
```

## 3.2 Análise Bivariada

### 1. Qual merchant que mais vendeu?

```python
# Agrupando por merchant
aux = df3[['merchant', 'date_imp_d']].groupby('merchant').count().sort_values(by='date_imp_d', ascending=False).reset_index()
sns.barplot( x='merchant', y='date_imp_d', data=aux );
```

### 2. Qual categoria mais vendida?

```python
# Top 5
aux1 = df3[['category_name', 'date_imp_d']].groupby('category_name').count().sort_values(by='date_imp_d', ascending=False).head().reset_index()
sns.barplot( x='category_name', y='date_imp_d', data=aux1 );
```

### 3. Qual a marca mais vendida?

```python
# Top 5
aux2 = df3[['brand', 'date_imp_d']].groupby('brand').count().sort_values(by='date_imp_d', ascending=False).head().reset_index()
sns.barplot( x='brand', y='date_imp_d', data=aux2 );
```
