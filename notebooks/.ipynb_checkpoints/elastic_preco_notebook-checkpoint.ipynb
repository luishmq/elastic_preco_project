# 5.0 Machine Learning Modelling

## 5.1 Unique example

### 5.1.1 Training linear regression

```python
x_price1 = x_price['Acer 15.6 Chromebook CB5-571-C4G4']
y_demand1 = y_demand['Acer 15.6 Chromebook CB5-571-C4G4']


X = sm.add_constant(x_price1)
model = sm.OLS(y_demand1, X)
results = model.fit()


results.summary()
```

### 5.1.2 Regression Visualization

```python
# Criar um gráfico de dispersão dos pontos de dados
plt.scatter(x_price1, y_demand1, label='Dados reais')

# Criar a linha de regressão com base nos resultados do modelo
plt.plot(x_price1, results.predict(X), label='Linha de Regressão', color='red')

# Adicionar rótulos e legendas ao gráfico
plt.xlabel('Preço do Acer Chromebook')
plt.ylabel('Demanda do Acer Chromebook')
plt.legend()

# Mostrar o gráfico
plt.show()
```

## 5.2 In general

### 5.2.1 Training linear regression

```python
results_values_laptop = {
    "name": [],
    "price_elasticity": [],
    "price_mean": [],
    "quantity_mean": [],
    "intercept": [],
    "slope": [],
    "rsquared": [],
    "p_value": []

}

df = pd.DataFrame()
for column in x_price.columns:
    #print(column)
    list_price = []
    list_demand = []
    column_points = []
    for i in range(len(x_price[column])):
        column_points.append({'x_price': x_price.reset_index(drop=True)[column][i],
                            'y_demand': y_demand.reset_index(drop=True)[column][i]})
        
        list_price.append(column_points[i]['x_price'])
        list_demand.append(column_points[i]['y_demand'])


    X = sm.add_constant(list_price)
    model = sm.OLS(list_demand, X)
    results = model.fit()

    if results.f_pvalue < 0.05:
        rsquared = results.rsquared
        pvalue = results.f_pvalue
        intercept, slope = results.params
        
        mean_price = np.round(np.mean(list_price),2)
        mean_quantity = np.round(np.mean(list_demand),2)

        price_elastity = slope*(mean_price/mean_quantity)

        results_values_laptop['name'].append(column)
        results_values_laptop['price_elasticity'].append(price_elastity)
        results_values_laptop['rsquared'].append(rsquared)
        results_values_laptop['p_value'].append(pvalue)
        results_values_laptop['intercept'].append(intercept)
        results_values_laptop['slope'].append(slope)
        results_values_laptop['price_mean'].append(mean_price)
        results_values_laptop['quantity_mean'].append(mean_quantity)
df_elastity = pd.DataFrame(results_values_laptop)
```

### 5.2.2 Regression Visualization

```python
# Lista dos nomes dos produtos
produtos = [
    '12 MacBook (Mid 2017, Silver)',
    'ASUS VivoBook Max X541SA 15.6Inch Laptop Intel Pentium 4GB Memory 500GB HD Matte IMR X541SA-PD0703X',
    'Acer 15.6 Chromebook CB5-571-C4G4',
    'Apple MacBook Pro MLUQ2LL/A 13.3 Notebook - Intel Core i5 Dual-core (2 Core) 2 GHz - 8 GB LPDDR3 - 256 GB SSD - Mac OS X 10.12 Sierra - 2560 x 1600 - In-plane Switching (IPS) Technology - Silver"',
    'Dell - Inspiron 15.6 Laptop - Intel Core i5 - 8GB Memory - NVIDIA GeForce GTX 1050 - 1TB + 8GB Hybrid Hard Drive - Red"',
    'Details About Apple Macbook Air 13.3 Laptop (early 2015) 2.2ghz Core I7 8gb 256gb"',
    'Lenovo - 100S-14IBR 14 Laptop - Intel Celeron - 2GB Memory - 32GB eMMC Flash Memory - Navy blue"'
]

# Defina as dimensões da grade
num_linhas = 2
num_colunas = 4

# Crie uma figura e uma grade de subplots
fig, axs = plt.subplots(num_linhas, num_colunas, figsize=(16, 8))

# Loop para criar gráficos de regressão para cada produto
for i, produto in enumerate(produtos):
    # Calcule as coordenadas da subplot na grade
    linha = i // num_colunas
    coluna = i % num_colunas

    # Selecione os dados para o produto atual
    x_price2 = x_price[produto]
    y_demand2 = y_demand[produto]

    # Adicione uma constante aos dados independentes
    X2 = sm.add_constant(x_price2)

    # Crie o modelo de regressão linear
    model2 = sm.OLS(y_demand2, X2)
    results2 = model2.fit()

    # Crie um gráfico de dispersão dos pontos de dados
    axs[linha, coluna].scatter(x_price2, y_demand2, label=f'Dados reais - {produto}')

    # Crie a linha de regressão com base nos resultados do modelo
    axs[linha, coluna].plot(x_price2, results2.predict(X2), label=f'Linha de Regressão - {produto}')

    # Adicione rótulos ao subplot
    axs[linha, coluna].set_xlabel('Preço dos Produtos')
    axs[linha, coluna].set_ylabel('Demanda dos Produtos')
    axs[linha, coluna].set_title(produto)
    

# Ajuste o layout para melhorar a legibilidade
plt.tight_layout()

# Mostrar os gráficos em uma grade
plt.show()
```

# 6.0 Business Performance

```python
result_faturamento = {
    'name': [],
    'faturamento_atual': [],
    'faturamento_reducao': [],
    'perda_faturamento': [],
    'faturamento_novo': [],
    'variacao_faturamento': [],
    'variacao_percentual': []
}

for i in range(len(df_elastity)):
    #print(i)
    preco_atual_medio = df_elastity['price_mean'][i] #x_price[df_elastity['name'][i]].mean()
    demanda_atual = y_demand[df_elastity['name'][i]].sum()

    reducao_preco = preco_atual_medio*0.9
    aumento_demanda = -0.1*df_elastity['price_elastity'][i]

    demanda_nova = aumento_demanda*demanda_atual

    faturamento_atual = round(preco_atual_medio*demanda_atual, 2)
    faturamento_novo = round(reducao_preco*demanda_nova, 2)

    faturamento_reducao = round(faturamento_atual*0.9, 2)

    perda_faturamento = round(faturamento_atual-faturamento_reducao, 2)

    variacao_faturamento = round(faturamento_novo-faturamento_atual ,2)

    variacao_percentual = round(((faturamento_novo-faturamento_atual)/faturamento_atual),2)

    # print('----------------------------------')
    # print(df_elastity['name'][i])
    # print(faturamento_atual)
    # print(faturamento_reducao)


    result_faturamento['name'].append(df_elastity['name'][i])
    result_faturamento['faturamento_atual'].append(faturamento_atual)
    result_faturamento['faturamento_reducao'].append(faturamento_reducao)
    result_faturamento['perda_faturamento'].append(perda_faturamento)
    result_faturamento['faturamento_novo'].append(faturamento_novo)
    result_faturamento['variacao_faturamento'].append(variacao_faturamento)
    result_faturamento['variacao_percentual'].append(variacao_percentual)

resultado = pd.DataFrame(result_faturamento)
resultado
```
